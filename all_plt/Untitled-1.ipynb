```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#rezise your graph(dpi specifies pixels per inch. when saving probably should use 300 if possible)
plt.figure(figsize=(5,3), dpi=100)

#line 1

#keyword argument notation
#plt.plot(x, y, label="2x", color="red", linewidth=2, marker=".", linestyle="--", markersize=10, markeredgecolor="blue")

#shorthand notation
#fmt = "[color][marker][line]"
plt.plot(x, y, 'b^--', label="2x")

#line num two

#select interval we want to plot points at
x2 = np.arange(0,4.5,0.5) #(start, stop, step)

#plot part of the graph as a line
plt.plot(x2[:6], x2[:6]**2, "r", label="x^2")

#plot remainder of the graph as a dot
plt.plot(x2[5:], x2[5:]**2, "r--")

#add a title(also an specific font and a font size)
plt.title('Our First Graph!', fontdict={"fontname": "Comic Sans MS", "fontsize": 20})

#x and y labels
plt.xlabel('X axis')
plt.ylabel('Y axis')

#X, Y axis tickmarks(scale of your graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

#add legend
plt.legend()

#save figure (dpi 300 is good when saving as a high resolution)
plt.savefig("mygraph.png", dpi=300)

#show plot
plt.show()
```

### Bar Chart

```python
labels = ["A", "B", "C"]
values = [1,4,2]

bars = plt.bar(labels, values)

#set patterns
patterns = ["/", "o", "*"]
for bars in bars:
    bars.set_hatch(patterns.pop(0))
#bars[0].set_hatch("//")
#bars[1].set_hatch("o")
#bars[2].set_hatch("*")

plt.figure(figsize=(6,4))

plt.show()
```
