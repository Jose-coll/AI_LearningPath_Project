```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
#(x, y, datos, organizacion con una key, tamaño, paleta de colores)
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size", palette="YlGnBu")
plt.show()
```

```python
#kde is a curve that shows the probability of the data on the graph
sns.histplot(tips["tip"], kde=True, bins=15)
```

```python
sns.displot(tips["tip"], kde=True, bins=15)
```

### Bar Plots

```python
sns.barplot(x="sex", y="tip", data=tips, palette="YlGnBu")
```

```python
sns.boxplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu")
```

```python
#the parameter dodge if it's true it will separate the hue, if it's false it will stay all together
sns.stripplot(x="day", y="tip", data=tips, hue="sex", dodge=True, palette="YlGnBu")
```

```python
#investigate in the documentation all the different kinds, they are freaking infinite
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", shade=True, cmap="YlGnBu")
#sns.jointplot(x="tip", y="total_bill", data=tips, kind="hex", cmap="YlGnBu")
#sns.jointplot(x="tip", y="total_bill", data=tips, kind="reg")
```

### Heat Map

```python
# Select only numeric columns from the titanic DataFrame
numeric_titanic = titanic.select_dtypes(include=['number'])

# Compute the correlation matrix and plot the heatmap
sns.heatmap(numeric_titanic.corr(), annot=True, cmap="YlGnBu")

#this type of graphs are about correlation, the darkest blue it is it is more correlated
#and the darkest yellow it is, it is less correlated
#you can also use as cmap which is usually the most common one, coolwarm, the coldest the color is the less correlated it is and the hottest the color is the more correlated it is
#probably one of the most helpful graphs 
```
